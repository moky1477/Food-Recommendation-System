```python
import re
from nltk.tokenize import word_tokenize

def clean_text(text):
    # Lowercase the text
    text = text.lower()
    
    # Remove special characters, numbers, specific units (excluding commas)
    text = re.sub(r'[^a-zA-Z\s,]', '', text)
    text = re.sub(r'\b(?:lb|tsp|tbsp|total|cup|cups|oz|teaspoons|teaspoon)\b', '', text)
    
    # Tokenize the text into words
    words = word_tokenize(text)
    
    # Join the words back into a cleaned string
    cleaned_text = ' '.join(words)
    
    return cleaned_text

# Input data as a single string
input_data = data["Ingredients"][0]

# Clean the data
cleaned_data = clean_text(input_data)

# Print the cleaned data
print(cleaned_data)
```

```python
data['Cleaned_Ingredients_Stem'] = data['Ingredients'].apply(clean_text)
```

### Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

### Lemmetization

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lm = WordNetLemmatizer()

def lemm(text):
    y = []
    
    for i in text.split():
        y.append(lm.lemmatize(i))
        
    return " ".join(y)
```

### Apply Stemming First

```python
data["Cleaned_Ingredients_Stem"] = data["Cleaned_Ingredients_Stem"].apply(stem)
```

### Text Vectorization

```python
# Text Vectorization using Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(data["Cleaned_Ingredients_Stem"]).toarray()
```

```python
vectors.shape
```

```python
len(cv.get_feature_names_out())
```

```python
# Cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(dish):
    dish_index = data[data['Title'] == dish].index[0]
    distances = similarity[dish_index]
    rec_list = []
    dish_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in dish_list:
        rec_list.append(data["Title"][i[0]])
        
    return rec_list
```

```python
from PIL import Image
import IPython.display as display

def recommend(dish):
    dish_index = data[data['Title'] == dish].index[0]
    distances = similarity[dish_index]
    dish_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i, dish_info in enumerate(dish_list, start=1):
        recommended_dish_index = dish_info[0]
        recommended_dish = data.iloc[recommended_dish_index]
        
        title = recommended_dish['Title']
        ingredients = recommended_dish['Cleaned_Ingredients']  # Assuming you have an 'Ingredients' column in your dataset
        image_name = recommended_dish['Image_Name']
        
        # Load and display the image
        image_path = f"Food Images/Food Images/{image_name}.jpg"  # Update the path to your image folder
        img = Image.open(image_path)
        display.display(img)
        
        print(f"{i}. Dish Name: {title}")
        print(f"   Ingredients: {ingredients}")
        print("\n")

# Example usage:
liked_dish = "Thanksgiving Mac and Cheese"
recommend(liked_dish)
```
